```python
img_folder = './SCUT-FBP5500_v2/Images/'
scores = pd.read_csv('SCUT-FBP5500_v2/train_test_files/All_labels.txt', 
                    sep=' ', names=['image_name', 'rating'])
scores['group'] = scores.image_name.apply(lambda name: name[:2])
```

```python
scores
```

```python
# most rating img
for img in scores.sort_values(by='rating', ascending=False).image_name.iloc[0:5]:
    print(img, ', score:', scores[scores.image_name==img].rating.iat[0])
    display(Image.open(os.path.join(img_folder, img)))
```

```python
# last rating img
for img in scores.sort_values(by='rating', ascending=True).image_name.iloc[:5]:
    print(img, ', score:', scores[scores.image_name==img].rating.iat[0])
    display(Image.open(os.path.join(img_folder, img)))
```

```python
scores.group.value_counts()
```

```python
# most rating img in each group
for group in scores.group.value_counts().keys():
    print('='*100)
    print('GROUP:', group)
    for img in scores[scores.group == group].sort_values(by='rating', ascending=False).image_name.iloc[0:5]:
        print(img, ', score:', scores[scores.image_name==img].rating.iat[0])
        print('-'*50)
        display(Image.open(os.path.join(img_folder, img)))
```

```python
# last rating img in each group
for group in scores.group.value_counts().keys():
    print('='*100)
    print('GROUP:', group)
    for img in scores[scores.group == group].sort_values(by='rating', ascending=True).image_name.iloc[0:5]:
        print(img, ', score:', scores[scores.image_name==img].rating.iat[0])
        print('-'*50)
        display(Image.open(os.path.join(img_folder, img)))
```

```python
import seaborn as sns
```

```python
# rating distribution
plt.figure()
sns.distplot(a=scores.rating, color='b')
plt.title('Rating distribution')
plt.show()
```

```python
# rating distribution
groups = ['AM', 'AF', 'CM', 'CF']
col = ['c', 'm', 'y', 'k']
plt.figure(figsize=(8,8))
for i in range(4):
    gr = groups[i]
    plt.subplot(2,2,i+1)
    sns.distplot(a=scores[scores.group==gr].rating, color=col[i])
    plt.title('Group: ' + gr)
    plt.xlabel('')
plt.suptitle('Rating distribution')    
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
dfa = scores[scores.group.apply(lambda gr: 'F' in gr)]
sns.distplot(a=dfa.rating, color='c')
plt.title('Female')
plt.xlabel('')

plt.subplot(2,2,2)
dfa = scores[scores.group.apply(lambda gr: 'M' in gr)]
sns.distplot(a=dfa.rating, color='m')
plt.title('Male')
plt.xlabel('')

plt.subplot(2,2,3)
dfa = scores[scores.group.apply(lambda gr: 'A' in gr)]
sns.distplot(a=dfa.rating, color='y')
plt.title('Asian')
plt.xlabel('')

plt.subplot(2,2,4)
dfa = scores[scores.group.apply(lambda gr: 'C' in gr)]
sns.distplot(a=dfa.rating, color='k')
plt.title('CaucAsian')
plt.xlabel('')

plt.suptitle('Rating distribution')    
plt.show()
```
